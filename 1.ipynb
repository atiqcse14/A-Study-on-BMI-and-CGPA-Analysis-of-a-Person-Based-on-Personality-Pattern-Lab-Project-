```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
D=df3.values


# Get the labelled set
c1=D[:20,:]; c2=D[51:80,:];  c3=D[91:100,:]
trainSet = np.concatenate((c1,c2,c3),axis=0)

# Get the testing set
c1 = D[21:50,:]; c2=D[81:90,:];  c3=D[100:,:]
testSet = np.concatenate((c1,c2,c3),axis=0)

print(trainSet.shape)
print(testSet.shape)

xTrain=trainSet[:,:-1]; yTrain=trainSet[:,-1]
xTest=testSet[:,:-1]; yTest=testSet[:,-1]

print(yTest)
```

```python

# create a knn classifier with K=3
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(xTrain, yTrain.astype(str))
```

```python


# Make predictions
yPred=clf.predict(xTest)
acc=accuracy_score(yTest.astype(str), yPred.astype(str))
print('Accuracy with 3 neighbours: ',acc)
```

```python

def plot_conf_mat(lTrue, lPred, title):
    """ A function for plotting the confusion matrix given true and predicted labels."""
    cm = confusion_matrix(lTrue.astype(str), lPred.astype(str))
    print(cm)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()  
```

```python
plot_conf_mat(yTest, yPred, 'K=3')
```
